## Import library

```python
import pandas as pd
import numpy as np
```

## Read data files

```python
Train = pd.read_csv('train.csv')
Test = pd.read_csv('test.csv')
Train
```

## Take only month in Attribute 1

```python
# train
Train['Attribute1']= pd.to_datetime(Train['Attribute1']) 
Train['Attribute1'] = pd.DatetimeIndex(Train['Attribute1']).month

# test
Test['Attribute1']= pd.to_datetime(Test['Attribute1']) 
Test['Attribute1'] = pd.DatetimeIndex(Test['Attribute1']).month
```

## Data preprocessing

### 處理NAN

```python
# 訓練data
Train = Train.dropna()
# 測試data
Test = Test.dropna()
```

### 處理字串欄位
字串轉數字

```python
from sklearn.preprocessing import LabelEncoder
```

```python
##.  Train Data ##
# 字串轉數字
gle = LabelEncoder()
Train['Attribute22'] = gle.fit_transform( Train['Attribute22'] )
Train['Attribute23'] = gle.fit_transform( Train['Attribute23'] )

## Test Data  ##
# 字串轉數字
Test['Attribute22'] = gle.fit_transform( Test['Attribute22'] )

# One-hot
Train = pd.get_dummies(Train)
Test = pd.get_dummies(Test)
Train
```

### Balance Data

```python
from sklearn.utils import resample

df_majority = Train[Train.Attribute23==0]
df_minority = Train[Train.Attribute23==1]

print(len(df_majority), len(df_minority))
print(Train['Attribute23'].value_counts())

# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=1225,     # to match minority class
                                 random_state=5659) # reproducible results
 
# Combine minority class with downsampled majority class
Train = pd.concat([df_majority_downsampled, df_minority])

# Display new class counts
print(Train['Attribute23'].value_counts())
```

### 打散資料

```python
from sklearn.utils import shuffle
Train = shuffle(Train)
```

### 切割 Data 與 Target

```python
Target = pd.DataFrame(Train['Attribute23'])
Data = Train.drop(columns = ['Attribute23'],axis = 0)
```

## Train Test split

```python
from sklearn.model_selection import train_test_split

data_train,data_test,target_train,target_test = train_test_split(Data, Target,test_size = 0.5)
print('Train Data Shape: ',data_train.shape, '\nTrain target Shape: ',target_train.shape)
print('Test Data Shape: ',data_test.shape, '\nTest target Shape: ',target_test.shape)
```

## Knn model

```python
from sklearn.neighbors import KNeighborsClassifier

# model the Knn
knn = KNeighborsClassifier(n_neighbors=30)
# knn.fit(data_train, target_train.values.ravel())
knn.fit(Data, Target.values.ravel())

# predict data_test result
print('Test Accurancy:', knn.score(data_test,target_test))
# choose_k_best(knn,Data, Target)
knn_pred = knn.predict(Test)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree

clf = DecisionTreeClassifier()
clf = clf.fit(Data, Target)

print('Test Accurancy:', clf.score(data_test,target_test))
# choose_k_best(clf,Data,Target)
clf_pred = clf.predict(Test)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(Data, Target.values.ravel())

print('Test Accurancy:', gnb.score(data_test,target_test)) 
# choose_k_best(gnb,Data,Target)
gnb_pred = gnb.predict(Test)
```

## Neural Network models

```python
from sklearn.neural_network import MLPClassifier
# nn = MLPClassifier(activation= 'logistic' ,solver= 'adam' ,alpha= 0.0001 ,learning_rate= 'adaptive' ,learning_rate_init= 0.001,verbose=100 ,max_iter= 500 )
nn = MLPClassifier(solver='adam', activation='logistic',alpha= 0.0001,learning_rate= 'adaptive' , hidden_layer_sizes=(50,50), random_state=1,max_iter=1000,verbose=10,learning_rate_init=0.001)
nn.fit(Data, Target.values.ravel())
nn_pred = nn.predict(Test)
print('Test Accurancy:', nn.score(data_test,target_test)) 
```
